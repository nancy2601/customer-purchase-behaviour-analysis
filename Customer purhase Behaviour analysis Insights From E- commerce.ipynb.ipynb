```python
import pandas as pd

# Load the pre-generated dataset
df = pd.read_csv("Correct_Ecommerce_Retail_Dataset.csv")

# Convert InvoiceDate column to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
```

```python
## üîç Step 4: Exploratory Data Analysis (EDA)

#We will now explore the dataset to uncover customer behavior patterns, product sales trends, and revenue breakdowns. 
#This will help generate insights to inform business strategy and optimize customer targeting.
```

```python
#total revenue 
total_revenue = df["TotalPrice"].sum()
print(f"üí∞ Total Revenue Generated: ‚Çπ{total_revenue:,.2f}")
```

```python
##Top 10 Selling Products (by Revenue)
top_products = (
    df.groupby("Description")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products.plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Top 10 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# revenue by country 
revenue_country = (
    df.groupby("Country")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
)

revenue_country.plot(kind='bar', figsize=(10,5), color='salmon')
plt.title("Revenue by Country")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# monthly sales trend
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_sales = df.groupby("Month")["TotalPrice"].sum()

monthly_sales.plot(kind='line', marker='o', figsize=(10,4), color='green')
plt.title("Monthly Sales Trend")
plt.ylabel("Revenue")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## no. of unique customers 
num_customers = df["CustomerID"].nunique()
print(f"üßç Unique Customers: {num_customers}")
```

```python
#Average Basket Size (Revenue per Transaction)
avg_basket = df.groupby("InvoiceNo")["TotalPrice"].sum().mean()
print(f"üõí Avg Basket Value per Transaction: ‚Çπ{avg_basket:,.2f}")
```
